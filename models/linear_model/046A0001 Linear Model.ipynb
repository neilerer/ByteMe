```python
import pandas as pd
import numpy as np
import os
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as sm
```

```python
df=pd.read_csv("Linear Model/clean_data/046A0001_clean.csv")
df.dtypes
```

```python
df['Cumulative Time Taken']=df['Cumulative Time Taken'].astype(int)
# df['Visibility (km)'].value_counts()
# df['Visibility (km)']=df['Visibility (km)'].astype(float)
```

```python
df=df.drop('Unnamed: 0',1)
```

```python
df_lin_reg=pd.DataFrame({'StopSeq':df['StopSeq'],
                           'MonToWed':((df.WeekDay==0)|(df.WeekDay==1)|(df.WeekDay==2))*1,
                          'ThuFri':((df.WeekDay==3)|(df.WeekDay==4))*1,
                          'Sat':(df.WeekDay==5)*1,'Sun':(df.WeekDay==6)*1,
                           'h7':(df.Hours==7)*1,'h8':(df.Hours==8)*1,'h9':(df.Hours==9)*1,
                           'h10':(df.Hours==10)*1,'h11':(df.Hours==11)*1,'h12':(df.Hours==12)*1,
                           'h13':(df.Hours==13)*1,'h14':(df.Hours==14)*1,'h15':(df.Hours==15)*1,
                           'h16':(df.Hours==16)*1,'h17':(df.Hours==17)*1,'h18':(df.Hours==18)*1,
                           'h19':(df.Hours==19)*1,'h20':(df.Hours==20)*1,'h21':(df.Hours==21)*1,
                           'h22':(df.Hours==22)*1,'h23':(df.Hours==23)*1,
                            'Temp':df['Temperature (C)'], 'Conditions':df['Conditions'],
                           'CTT':df['Cumulative Time Taken']})
cols=['StopSeq','MonToWed','ThuFri','Sat','Sun','h7','h8','h9','h10','h11','h12','h13','h14','h15','h16','h17','h18','h19','h20','h21','h22','h23','Temp','Conditions','CTT']
df_lin_reg=df_lin_reg[cols]
```

```python
df_lin_reg.head(10)
```

```python
lm = sm.ols(formula="CTT~StopSeq+MonToWed+ThuFri+Sat+Sun+h7+h8+h9+h10+h11+h12+h13+h14+h15+h16+h17+h18+h19+h20+h21+h22+h23+Temp+Conditions", data=df_lin_reg).fit()
```

```python
lm.params
```

```python
lm.summary()
```

```python
lm = sm.ols(formula="CTT~StopSeq+MonToWed+ThuFri+Sat+Sun+h7+h8+h9+h10+h11+h12+h13+h14+h15+h16+h17+h18+h19+h20+h21+h22+h23+Temp+Conditions", data=df_lin_reg).fit()
```

```python
lm_predictions=lm.predict(df_lin_reg)
actual=df_lin_reg['CTT']
```

```python
a=pd.DataFrame({'Result':lm_predictions,'Actual':actual,'Error':abs(actual/lm_predictions/100)})
np.mean(a['Error'])
```

```python
train, test = train_test_split(df_lin_reg, test_size = 0.3)
```

```python
lm_train=lm = sm.ols(formula="CTT~StopSeq+MonToWed+ThuFri+Sat+Sun+h7+h8+h9+h10+h11+h12+h13+h14+h15+h16+h17+h18+h19+h20+h21+h22+h23+Temp+Conditions", data=train).fit()
```

```python
test_actual=test['CTT']
```

```python
lm_train_predictions=lm_train.predict(test)
```

```python
a=pd.DataFrame({'Result':lm_train_predictions,'Actual':test_actual,'Accuracy':abs(test_actual/lm_train_predictions)})
print(np.mean(a['Accuracy']))
```
